## Qual é a Receita total?

```python
df["Valor Venda"].sum()
```

## Criando uma nova coluna com o "Custo Total"

```python
df["Custo Total"] = df["Custo Unitário"].mul(df["Quantidade"])
```

## Verificando o resultado utilizando uma amostra com método sample()

```python
df.sample()
```

## Retornando o "Custo Total" arredondado - Qual é o "Custo Total"?

```python
round(df["Custo Total"].sum(), 2)
```

## Criando a coluna 'Lucro Total' (Receita que é o 'Valor Venda' - 'Custo Total')

```python
df["Lucro Total"] = df["Valor Venda"] - df["Custo Total"]
```

```python
df.head()
```

## Total de Lucro arredondado

```python
# Podemos fazer como no exemplo acima, ou podemos armazenar em uma variável
lucro_total = round(df["Lucro Total"].sum(), 2)

print(lucro_total)
```

## Criando uma coluna com 'Tempo_envio' -> Total de dias para enviar o produto

```python
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head()
```

## Alterando a coluna 'Tempo_envio' em minutos 

Agora, queremos saber a média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo_envio em minutos

```python
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days # ou seja, extraia para mim apenas os dias
```

```python
df.head()
```

## Verificando o tipo de coluna 'Tempo_envio' primeiro

```python
df["Tempo_envio"].dtype
```

## Média do tempo de envio por 'Marca'

```python
df.groupby("Marca")["Tempo_envio"].mean()
```

## Situação 4 - E se for necessário saber o 'Lucro Total' por ano e 'Marca'?

### Agrupar por ano e marca:

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro Total"].sum()
```

Os valores estão quebrados. Vamos formata-los

### Formatação de números

```python
pd.options.display.float_format = '{:20,.3f}'.format
```

Ao trazermos o código novamente:

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro Total"].sum()
```

Os valores estarão corrigidos.

## ***Resetando o index - Formatação e retornando uma nova tabela com os valores tidos acima***

```python
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro Total"].sum().reset_index()
lucro_ano
```
