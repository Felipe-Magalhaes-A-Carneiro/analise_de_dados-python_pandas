## ***Quais são os valores únicos de uma coluna?***

### unique()

Quando utilizado com o 'acesso direto' podemos trazer apenas os valores únicos que existe em uma coluna.

```python
df["Continente"].unique()
```

perceba que ele traz um 'array'. O que é um array?

Se você ouvir alguém falando de "arrays em Python" no contexto de Ciência de Dados ou Inteligência Artificial, eles estão falando da biblioteca NumPy. Esses arrays são incrivelmente rápidos e permitem cálculos matemáticos complexos em massa.

Exemplo de código com este uso:

Criando um array do NumPy:

import numpy as np

arr = np.array([1, 2, 3])
print(arr * 2)  # Multiplica todos os itens por 2 de uma vez: [2, 4, 6]

## ***Agrupamento de Dados***

Agrupamento de dados é muito comum em Banco de Dados. Aqui mostraremos como fazer o mesmo com o Pandas.

### groupby()

"Quantos paises nós temos para cada continente?"

```python
# 'agrupo' pelo continente, 'encontre pelo' pais e 'retorne' a soma de paises que cada um tem com nunique() 
df.groupby("Continente")["Pais"].nunique()
```

### groupby() + mean()

"Para cada ano, qual era a 'expecitativa média de vida' de cada país? "

```python
# 'agrupo' pelo ano, 'encontre pela' a exp. de vida e 'retorne' a média 
df.groupby("Ano")["Exp. de vida"].mean()
```
