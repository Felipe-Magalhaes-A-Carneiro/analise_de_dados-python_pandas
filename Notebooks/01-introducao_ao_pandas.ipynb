## ***Acessando os primeiros dados do arquvio:***

### head()

Por padrão, ele retornando as 5 primeiras linhas do arquivo.

Obs.: Como é por index, ele retorna do 0 ao 4, totalizando os 5 primeiros index.

```python
df.head()
```

Caso queira que ele retorne outro número de linhas específico, coloque o número desejado dentro dos seus parênteses, o que o fará ser um parâmetro

```python
df.head(11)
```

--------------------------------------------------------------------------------------------------

## ***Acessando os últimos dados do arquvio:***

### tail()

Por padrão, ele retornando as 5 últimas linhas do arquivo.

O mesmo ocorre com o head(), caso queira um número específico para retornar, coloque este dentro dos seus parâmetros.

Obs.: Como é por index, ele retorna do 0 ao 4, totalizando os 5 últimos index.

```python
df.tail(11)
```

# ***Formatando o arquivo***

--------------------------------------------------------------------------------------------------

## ***Formatando as colunas***

Já que as colunas estão em inglês, vamos alterar apenas elas para português.

Utilizaremos o método rename() e em seu parâmetro o 'columns' que deve aceitar os novos valores por meio da Estrutura de Dados 'Dicionário'.

*nome em inglês * : * nome em portugues *

```python
df.rename(columns={"country": "Pais", "continent": "Continente", "year": "Ano", "lifeExp": "Exp. de vida", "pop": "Pop. Total", "gdpPercap": "PIB"})
```

--------------------------------------------------------------------------------------------------

# ***Salvando as alterações no 'df' (dataframe)***

```python
# Simplismente atribua o 'df' ao códgigo de formatação feita pro você

df = df.rename(columns={"country": "Pais", "continent": "Continente", "year": "Ano", "lifeExp": "Exp. de vida", "pop": "Pop. Total", "gdpPercap": "PIB"})
```

## ***Retornando - Número de linhas e colunas que um arquivo possuí***

### shape

Retorna uma tupla com os valores que o arquivo dispõe.

(* número de linhas *, * número de colunas *)

Obs.: Perceba que o 'shape' não precisa dos parênteses '()' depois dele.

```python
df.shape
```

--------------------------------------------------------------------------------------------------

## ***Retornando - Apenas os nomes das colunas***

### columns

Retorna uma tupla com uma lista com os valores que o arquivo dispõe e o seu tipo por último.

Obs.: Perceba que o 'columns' não precisa dos parênteses '()' depois dele.

```python
df.columns
```

## ***Retornando - Tipo de dado em cada coluna***

### dtypes

Retorna uma tabela com o nome de cada coluna e, ao lado, o seu tipo de cado que ele armazena.

Obs.: Perceba que o 'dtypes' não precisa dos parênteses '()' depois dele.

```python
df.dtypes
```
